```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_ind

categories = ['Traditional 2D', 'Our vizualization']

def analyze(data, title, ylabel):
    # Perform t-test
    t_stat, p_value = ttest_ind(data[0], data[1])
    if p_value < 0.05:
        result = f"The difference between the two categories is statistically significant (p < 0.05). t = {t_stat:.2f}, p = {p_value:.2f}"
    else:
        result = f"The difference between the two categories is not statistically significant (p >= 0.05). t = {t_stat:.2f}, p = {p_value:.2f}"

    print(result)

    # Plot the data
    plt.figure(figsize=(5, 5))
    values = [sum(v) / len(v) for v in data]
    errors = [np.std(v, ddof=1) / np.sqrt(len(v)) for v in data]  # Use ddof=1 for sample SD

    plt.bar(categories, values)
    plt.errorbar(categories, values, yerr=errors, fmt='o', color='black', capsize=5)

    plt.title(title)
    plt.ylabel(ylabel)

    colors = ['tab:orange' if i in [0, 3] else 'tab:blue' for i in range(len(categories))]
    plt.bar(categories, values, color=colors)
    plt.margins(x=0.2, y=0.2)

    plt.show()
```

```python
# Data: Number of insights
print('Number of insights')
data = [[4, 5, 3, 4, 5, 2, 3, 5, 4, 3],
        [3, 4, 3, 4, 5, 3, 2, 5, 3, 4]]
analyze(data, 'How many insights did participants list?', 'Number of insights')
```

```python
# Data: Time spent on each visualization
print('Time spent on each visualization')
data = [[10, 15, 10, 20, 30, 10, 7, 10, 20, 15],
    [40, 50, 30, 40, 60, 20, 30, 30, 35, 40]]
analyze(data, 'How much time spent on each visualization?', 'Time spent (seconds)')
```

```python
# Data: Rate the quality of the visualization
print('Rate the quality of the visualization')
data = [[3, 2, 3, 4, 4, 3, 3, 4, 3, 3],
        [6, 6, 5, 6, 7, 6, 5, 5, 6, 6]]
analyze(data, 'Rate the quality of the visualization\n(1 = low, 7 = high)', 'Rating')
```
