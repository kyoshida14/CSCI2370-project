```python
# THIS MAKES YOU DAYX_DATA.CSV

import pandas as pd
import numpy as np

import os
# Load the CSV file into a pandas DataFrame
# Replace 'your_data.csv' with the actual path to your CSV file
full_data = pd.read_csv('SimData2_HumansOnly.csv', index_col=0)

# List of agent types to select (subset of columns)
selected_agent_types = ['S_H','E_H','E_TH','E_NT','I_H','I_QH','D_H']  # Example, adjust to your needs

# Initialize an empty dictionary to store the individual day dataframes
day_datasets = {}

# Define the folder where you want to save the CSV files
output_folder = 'dayx_datas'  # Replace with your desired folder name

# Create the folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Iterate through each row (representing a day)
for day in full_data.index:
    # Select the relevant agent types and multiply by 10^2
    day_data = (full_data.loc[day, selected_agent_types] * 10**(2)).astype(int)

    # Create a dataset for that day, called 'dayx_data' (e.g., 'Day1_data')
    day_datasets[f'{day}_data'] = day_data.to_frame().T  # Convert to DataFrame with one row

    # Save the dataset for the day as a CSV file in the specified folder
    day_filename = os.path.join(output_folder, f"{day}_data.csv")
    day_data.to_frame().T.to_csv(day_filename, index=False)

!zip -r dayx_datas.zip dayx_datas
from google.colab import files
files.download('dayx_datas.zip')

# Save the dataset for the day as a CSV file in the specified folder
day_filename = os.path.join(output_folder, f"{day}_data.csv")
day_data.to_frame().T.to_csv(day_filename)
```

```python
import pandas as pd
import numpy as np
import os
from glob import glob

# Define input and output folders
input_folder = 'dayx_datas'  # Folder containing dayx_data files
output_folder = 'dayx_coords'  # Folder to save dayx_coords files

# Color mapping for agent types
state_var_colors = {
    "S_H": "#F1C40F",  # [241, 196, 15]
    "E_H": "#EF5F33",       # [239, 95, 51]
    "E_TH": "#E26A6A" ,  # [226, 106, 106]
    "E_NT": "#CF366C",   # [207, 54, 108]
    "I_H": "#3498DB",          # [52, 152, 219]
    "I_QH": "#14C7DE",          # [20, 199, 222]
    "D_H": "#515C5D"          # [81, 92, 93]
}


# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Get a list of all dayx_data CSV files
input_files = glob(os.path.join(input_folder, '*.csv'))

# Process each file
for input_file in input_files:
    # Extract the "day" from the filename for naming the output
    day_name = os.path.basename(input_file).replace('.csv', '')

    # Load the population counts CSV
    population_data = pd.read_csv(input_file)

    # Initialize an empty list to store data for this day
    all_data = []
    print(day_name)
    #delete all Unnamed Columns in a single code of line using regex
    population_data.drop(population_data.filter(regex="Unnamed"),axis=1, inplace=True)
    # For each agent type, get the count and generate that many random coordinates
    for agent_type in population_data.columns:
        print(agent_type)

        count = population_data.iloc[0][agent_type]  # Get count for the agent type

        # Generate random coordinates
        x_coords = np.random.rand(int(count)) * 100  # Scale as needed
        y_coords = 0
        z_coords = np.random.rand(int(count)) * 100

        color = state_var_colors[agent_type]
        print(state_var_colors[agent_type])
        # Create a DataFrame for this agent type with the random coordinates
        agent_df = pd.DataFrame({
            'x': x_coords,
            'y': y_coords,
            'z': z_coords,
            'agent_type': agent_type,
            'color': color
        })

        # Append to the master list
        all_data.append(agent_df)

    # Concatenate all agent DataFrames into one for this day
    final_df = pd.concat(all_data, ignore_index=True)

    # Save the generated coordinates to the output folder
    output_file = os.path.join(output_folder, f"{day_name}_coords.csv")
    final_df.to_csv(output_file, index=False)

# Zip the output folder for download
!zip -r dayx_coords.zip dayx_coords
from google.colab import files
files.download('dayx_coords.zip')
```
